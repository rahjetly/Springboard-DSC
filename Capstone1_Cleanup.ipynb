```python
#Now I want to map the stats from the 'fighter_details' DF to the 'fight_data' DF where there 
#will be new columns created based on whether it is R fighter details or B fighter details
#I want all of the columns from 'fighter_details' to be added on

all_data = pd.merge(fight_data, fighter_details, how = 'left', left_on = 'R_fighter', right_on = 'fighter_name')
```

```python
all_data.rename(columns = {'Height':'R_Height', 'Weight':'R_Weight', 'Reach':'R_Reach', 'Stance':'R_Stance', 'DOB':'R_DOB'}, inplace=True)
```

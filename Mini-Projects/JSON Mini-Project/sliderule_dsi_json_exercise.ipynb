## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

### Loading in the full file

### Question 1 - Find the 10 countries with most projects

```python
#group by countries and print out 10 countries with the most projects
country_df = full_file.groupby('countryshortname').id.nunique().sort_values(ascending = False)
country_df.head(10)
```

### Question 2 - Find the top 10 major project themes

```python
full_string = json.load((open('data/world_bank_projects.json')))
```

```python
#Use JSON Normalize to print out top 10 most common project themes
pt = json_normalize(full_string, 'mjtheme_namecode', errors='ignore').groupby('name').size().sort_values(ascending = False)
pt.head(11)
```

### Question 3 - There are a significant amount of blanks so we need to fill in and then re-do the above list

```python
print(full_file['mjtheme_namecode'].head(30))
```

```python
import numpy as np
```

```python
#Create the same DF as df_pt, set empty values to NaN, then replace all NaN values by doing a forward fill
pt2 = json_normalize(full_string, 'mjtheme_namecode', errors='ignore')
pt2 = pd.DataFrame(pt2.replace('', np.NaN))
```

```python
pt2.head()
```

#### Get a dictionary of what code maps to which name and then merge the DFs to fill in the blanks

```python
#create new DF. Drop NaNs and duplicates.
pt3 = pt2.dropna()
pt3 = pt3.drop_duplicates()
pt3
```

```python
pt4 = pt3.merge(pt2, how='outer')
pt4.head(20)
```

```python
pt5 = pt4.groupby('code')
pt5.head(10)
```

```python
#Fill in blank values using forward fill as the DF is grouped by the appropriate code
pt5 = pt5.fillna(method='ffill')
```

```python
#check shape before dropping NaN values
pt4.shape
```

```python
#drop NaN values
pt5.dropna()
```

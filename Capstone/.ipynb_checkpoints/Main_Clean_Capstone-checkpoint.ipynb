### Modify the newly combined DF

Will start by removing the '%' symbol and turning strings to a float.

```python
def rm_pct(x):
    return float(x.strip('%'))/100
```

```python
stacked_data['takdown_pct'] = stacked_data['takdown_pct'].apply(rm_pct)
stacked_data['sig_strike_pct'] = stacked_data['sig_strike_pct'].apply(rm_pct)
```

Verifying that the function did its job.

```python
stacked_data['takdown_pct'].head()
```

```python
stacked_data['sig_strike_pct'].head()
```

It worked. Now we can perform rolling operations on select columns after we transform height, weight, and reach to numeric.

```python
def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = str(ht).split("' ")
    ft_ = float(ht_[0])
    in_ = float(ht_[1].replace("\"",""))
    return (12*ft_) + in_
```

```python
stacked_data['height'] = parse_ht(stacked_data['height'])
```

```python
type(stacked_data['height'][0])
```

```python
def rm_lb(x):
    return float(str(x).strip(' lbs.'))
```

```python
stacked_data['weight'] = stacked_data['weight'].apply(rm_lb)
```

```python
stacked_data['weight'].head()
```

```python
def rm_reach(x):
    return float(str(x).strip('"'))
```

```python
stacked_data['reach'].apply(rm_reach).head()
```

### Perform a rolling mean

My end goal is to have each row show the averages for a certain metric up until the current fight so there needs to be a shift such that current fight data isn't factored into the average. 

The new DF being created with the shift will only have columns that are needed for the rolling average.

```python
shift_df = stacked_data[['fighter_name','knock_down','sig_strike_pct','takdown_pct','submission_att','pass','rev']]\
    .groupby(['fighter_name'])\
    .shift(periods = 1, axis = 0)
```

Verify that it worked.

```python
shift_df.tail()
```

Now will bring this back in with the original DF so it shows stats up to current fight when I do the rolling mean.

```python
df4= pd.concat([stacked_data,shift_df], axis = 1)
```

Verify that this was brought in correctly.

```python
df4.head()
```

#### Time to do the rolling mean

```python
df5 = df4\
    .groupby('fighter_name')\
    .apply(lambda x: x.rolling(window=3, min_periods = 1, center=False, on='date', closed='right')[['knock_down','sig_strike_pct','takdown_pct','submission_att','pass','rev']]\
    .mean())
```

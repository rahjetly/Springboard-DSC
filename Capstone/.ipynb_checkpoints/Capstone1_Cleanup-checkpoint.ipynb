Now that DFs have been separated, going to stitch them together, do a group by for fighter name, then do a rolling mean of last 5 fights

```python
stacked_data = pd.concat([red_df, blue_df], axis=0)
```

```python
stacked_data = stacked_data.sort_values(['fighter_name', 'date'])
```

```python
stacked_data.info()
```

```python
stacked_data.reset_index(drop=True, inplace = True)
```

```python
#Trying to remove the % so I can turn certain columns to floats
def rm_pct(x):
    return float(x.strip('%'))/100
```

```python
stacked_data['takdown_pct'] = stacked_data['takdown_pct'].apply(rm_pct)
stacked_data['sig_strike_pct'] = stacked_data['sig_strike_pct'].apply(rm_pct)
```

```python
## NOW go ahead and change some of the above datatypes to float values then we will go in and implement the 
#rolling mean on the columns we change dtype for. Then will add in more features by broadening the DF with
#column values that can be separated
stacked_data = stacked_data.astype({'knock_down': 'float64', 'sig_strike_pct':'float64'})
```

```python
np.dtype(stacked_data['knock_down'])
```

```python
#shifting data
shift_df = stacked_data[['fighter_name','knock_down','sig_strike_pct','takdown_pct','submission_att','pass','rev']]\
    .groupby(['fighter_name'])\
    .shift(periods = 1, axis = 0)
```

```python
df4= pd.concat([stacked_data,shift_df], axis = 1)
```

```python
df4.head()
```

```python
list(df4.columns)
```

```python
#rename the columns
df4.columns=['fighter_name',
 'knock_down1',
 'sig_strikes',
 'sig_strike_pct1',
 'total_strikes',
 'takedown',
 'takdown_pct1',
 'submission_att1',
 'pass1',
 'rev1',
 'head',
 'body',
 'leg',
 'distance',
 'clinch',
 'ground',
 'height',
 'weight',
 'reach',
 'stance',
 'DOB',
 'date',
 'Winner',
 'win_by',
 'last_round',
 'Fight_type',
 'knock_down',
 'sig_strike_pct',
 'takdown_pct',
 'submission_att',
 'pass',
 'rev']
```

```python
#drop the initial columns
cols = [1,3,6,7,8,9]
df4.drop(df4.columns[cols], axis=1, inplace = True)
```

```python
#verify the drop occurred
df4.head()
```

```python
#Lets do the roll
df5 = df4\
    .groupby('fighter_name')\
    .apply(lambda x: x.rolling(window=3, center=False, on='date', closed='right')[['knock_down','sig_strike_pct','takdown_pct','submission_att','pass','rev']]\
    .mean())
```

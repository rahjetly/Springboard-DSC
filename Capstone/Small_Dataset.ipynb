### Feature to Feature Evaluation

Now I want to see if there is a positive correlation btwn any of the values so going to do a heatmap. The heatmap will be very cluttered so I'll focus on a correlation (pos or neg of >.5) to make it more digestable.

```python
corr = small_df.corr()
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(corr[(corr >= .5) | (corr <= -.5)], square = True, vmax = 1.0, vmin = -1.0, annot = True, annot_kws = {"size":8})
```

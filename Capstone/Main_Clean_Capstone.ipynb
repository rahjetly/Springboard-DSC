### Begin merging DFs

Starting to bring in the Red Fighter data first.

```python
part_data = pd.merge(fight_data, fighter_details, how = 'left', left_on = 'R_fighter', right_on = 'fighter_name')
```

```python
part_data.head()
```

```python
part_data.rename(columns = {'fighter_name':'Red_fighter','Height':'R_Height', 'Weight':'R_Weight', 'Reach':'R_Reach', 'Stance':'R_Stance', 'DOB':'R_DOB'}, inplace=True)
```

```python
part_data.head()
```

Bringing in Blue Fighter data now.

```python
all_data = pd.merge(part_data, fighter_details, how = 'left', left_on = 'B_fighter', right_on = 'fighter_name')
```

```python
all_data.head()
```

```python
all_data.rename(columns = {'fighter_name':'Blue_fighter','Height':'B_Height', 'Weight':'B_Weight', 'Reach':'B_Reach', 'Stance':'B_Stance', 'DOB':'B_DOB'}, inplace=True)
```

```python
all_data.head()
```

#### Change some value types and add in necessary columns in the new DF now.

Change "date" column to datetime type so I can sort later.

```python
all_data['date'] = all_data['date'].astype('datetime64[ns]') 
```

Now going to change the perspective of the problem to see if the Red Fighter won. IE: Introduce a column that solely tells us if the Red Fighter won.

```python
all_data['R_won'] = all_data['R_fighter']==all_data['Winner']
```
